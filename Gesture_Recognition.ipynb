## Generator
This is one of the most important part of the code. The overall structure of the generator has been given. In the generator, you are going to preprocess the images as you have images of 2 different dimensions as well as create a batch of video frames. You have to experiment with `img_idx`, `y`,`z` and normalization such that you get high accuracy.

```python
# Defining custom data generator

frames = 18 # number of frames, considering ~50% of total number of frames available in each of the ~10 fps video
width = 120 # image width
height = 120 # image height
channels = 3 # number of channels in each frame/image

def generator(source_path, folder_list, batch_size, sequences):
    print( 'Source path = ', source_path, '; batch size =', batch_size)
    img_idx = [frames for frames in range(6,frames+6)] #list of images to be selected in sequence for each video, maximum information captured during the intermediate frames
    
    while True:
        t = np.random.permutation(folder_list) #to shuffle the CSV files
        num_batches = math.floor(sequences/batch_size) #to calculate the number of batches
        
        for batch in range(num_batches): # we iterate over the number of batches
            batch_data = np.zeros((batch_size,frames,width,height,channels)) # frames is the number of images you use for each video, (width,height) is the final size of the input images and 3 is the number of channels RGB
            batch_labels = np.zeros((batch_size,5)) # batch_labels is the one hot representation of the output
            
            for folder in range(batch_size): # iterate over the batch_size
                imgs = os.listdir(source_path+'/'+ t[folder + (batch*batch_size)].split(';')[0]) # read all the images in the folder
                imgs.sort(reverse=True)
                
                for idx,item in enumerate(img_idx): #  Iterate the frames/images of a folder to read them in
                    image = imread(source_path+'/'+ t[folder + (batch*batch_size)].strip().split(';')[0]+'/'+imgs[item]).astype(np.float32)
                    
                    #crop the images and resize them. Note that the images are of 2 different shapes namely, 360 x 360 and 160 x 120
                    #and the conv3D will throw error if the inputs in a batch have different shapes

                    image = imresize(image,(width,height)) #resized images, cropping not carried since, affects accuracy
                    
                    batch_data[folder,idx,:,:,channels-3] = (image[:,:,channels-3] - np.min(image[:,:,channels-3]))/(np.max(image[:,:,channels-3])- np.min(image[:,:,channels-3])) #normalise using Min-Max, and feed in the image
                    batch_data[folder,idx,:,:,channels-2] = (image[:,:,channels-2] - np.min(image[:,:,channels-2]))/(np.max(image[:,:,channels-2])- np.min(image[:,:,channels-2])) #normalise using Min-Max, and feed in the image
                    batch_data[folder,idx,:,:,channels-1] = (image[:,:,channels-1] - np.min(image[:,:,channels-1]))/(np.max(image[:,:,channels-1])- np.min(image[:,:,channels-1])) #normalise using Min-Max, and feed in the image
                    
                batch_labels[folder, int(t[folder + (batch*batch_size)].strip().split(';')[2])] = 1
                
            yield batch_data, batch_labels #you yield the batch_data and the batch_labels, remember what does yield do

         # write the code for the remaining data points which are left after full batches
        rem_batch = num_batches*batch_size
        for rem_folder in range(batch_size): # iterate over the batch_size
            imgs = os.listdir(source_path+'/'+ t[(rem_folder + rem_batch) % sequences].split(';')[0]) # read all the images in the folder
            imgs.sort(reverse=True)
            
            for idx,item in enumerate(img_idx): #  Iterate over the frames/images of a folder to read them in
                    image = imread(source_path+'/'+ t[(rem_folder + rem_batch) % sequences].strip().split(';')[0]+'/'+imgs[item]).astype(np.float32)
                    
                    #crop the images and resize them. Note that the images are of 2 different shapes namely, 360 x 360 and 160 x 120
                    #and the conv3D will throw error if the inputs in a batch have different shapes

                    image = imresize(image, (width, height)) #resized images, cropping not carried out since, affects accuracy
                    
                    batch_data[folder,idx,:,:,channels-3] = (image[:,:,channels-3] - np.min(image[:,:,channels-3]))/(np.max(image[:,:,channels-3])- np.min(image[:,:,channels-3])) #normalise using Min-Max, and feed in the image
                    batch_data[folder,idx,:,:,channels-2] = (image[:,:,channels-2] - np.min(image[:,:,channels-2]))/(np.max(image[:,:,channels-2])- np.min(image[:,:,channels-2])) #normalise using Min-Max, and feed in the image
                    batch_data[folder,idx,:,:,channels-1] = (image[:,:,channels-1] - np.min(image[:,:,channels-1]))/(np.max(image[:,:,channels-1])- np.min(image[:,:,channels-1])) #normalise using Min-Max, and feed in the image
            
            batch_labels[rem_folder, int( t[(rem_folder + rem_batch) % sequences].strip().split(';')[2])] = 1
           
        yield batch_data, batch_labels #you yield the batch_data and the batch_labels, remember what does yield do
```

Note here that a video is represented above in the generator as (number of images, height, width, number of channels). Take this into consideration while creating the model architecture.

Images of both dimensions have been resized by keeping its aspect ratio. However, cropping is not carried out, as it affects model performance.

## Model
Here you make the model using different functionalities that Keras provides. Remember to use `Conv3D` and `MaxPooling3D` and not `Conv2D` and `Maxpooling2D` for a 3D convolution model. You would want to use `TimeDistributed` while building a `Conv2D + RNN` model. Also remember that the last layer is the softmax. Design the network in such a way that the model is able to give good accuracy on the least number of parameters so that it can fit in the memory of the webcam.

```python
# Importing necessary packages

from tensorflow.keras import layers
from keras.models import Sequential, Model
from keras.layers import Dense, Dropout, SimpleRNN, LSTM, GRU, Flatten, TimeDistributed, Flatten, BatchNormalization, Activation
from keras.layers.convolutional import Conv2D, Conv3D, MaxPooling2D, MaxPooling3D
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from keras import optimizers
from tensorflow.keras.regularizers import l2
```

```python
# Number of classes

classes = 5
print('Number of classes for softmax are',classes)
```

Below are the final experiments. Kindly refer to the seperate PDF document attached to see the training & model architecture of the initial experiments.

**Experiement-1**

```python
#write your model here

base_model = keras.applications.InceptionV3(include_top=False, weights='imagenet', input_shape=input_shape_tl)

model = Sequential()

model.add(TimeDistributed(base_model, input_shape=input_shape))

model.add(TimeDistributed(Flatten()))

model.add(SimpleRNN(32))

model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(32))
model.add(Activation('relu'))

model.add(Dropout(0.25))

model.add(Dense(5))
model.add(Activation('softmax'))
```

When you print the `summary` of the model, you'll see the total number of parameters you have to train.

```python
# Model summary

model.summary()
```

```python
# Choosing the optimizer with learning rate

learning_rate = 0.05

optimizer = tf.keras.optimizers.Adagrad(learning_rate=learning_rate)

print('# optimizer =', optimizer)
print('learning rate =', learning_rate)
```

Now that you have written the model, the next step is to `compile` the model.

```python
# Compiling our model

model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

We want to save our model, along with its learning rate, its loss & accuracy during both training and validation, for every epoch. To do this, we create checkpoints & callbacks, which is also useful for remembering the gradient updates. 

```python
# Creating checkpoints & callbacks

model_name = 'model_init' + '_' + str(curr_dt_time).replace(' ','').replace(':','_') + '/'
    
if not os.path.exists(model_name):
    os.mkdir(model_name)
        
filepath = model_name + 'model-{epoch:05d}-{loss:.5f}-{categorical_accuracy:.5f}-{val_loss:.5f}-{val_categorical_accuracy:.5f}.h5'

checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=False, save_weights_only=False, mode='auto', period=1)
```

We want our model to adjust its learning rate according to the validation loss, so that our model neither oscillates about the global minima nor get stuck in the local minima. So, we reduce our learning rate when there are no gradient updates consecutively for 2 epochs, to have a faster convergence. To do this, we write the `ReduceLROnPlateau` code.

```python
# Reduce the learning rate by what factor if plateaued

red_factor = 0.75

LR = ReduceLROnPlateau(monitor='val_loss', factor=red_factor, patience=2, verbose=1) #write the REducelronplateau code here
callbacks_list = [checkpoint, LR]

print('LR Reduction factor if val_loss plateaued =',red_factor)
```

Changed the argument "period" to "save_freq" in *checkpoint* due to below warning:

    WARNING:tensorflow: 'period' argument is deprecated. Please use 'save_freq' to specify the frequency in number of batches seen.

Changed the argument "period" back  in *checkpoint* after below error:

    KeyError: 'Failed to format this callback filepath: "model_init_2022-04-2916_17_49.011933/model-{epoch:05d}-{loss:.5f}-{categorical_accuracy:.5f}-{val_loss:.5f}-{val_categorical_accuracy:.5f}.h5". Reason: \'val_loss\''

The `steps_per_epoch` and `validation_steps` are used by `fit_generator` to decide the number of next() calls it need to make.

```python
# Calculate number of steps per epoch & validation steps based on number of training sequences & batch size

if (num_train_sequences%batch_size) == 0:
    steps_per_epoch = int(num_train_sequences/batch_size)
else:
    steps_per_epoch = (num_train_sequences//batch_size) + 1

if (num_val_sequences%batch_size) == 0:
    validation_steps = int(num_val_sequences/batch_size)
else:
    validation_steps = (num_val_sequences//batch_size) + 1
```

```python
# Number of steps per epoch & validation steps

print('steps per epoch =',steps_per_epoch)
print('validation steps =',validation_steps)
```

```python
# Choosing the number of epochs

num_epochs = 10
print ('# epochs =', num_epochs)
```

```python
# Importing necessary library

import time
```

Let us now fit the model. This will start training the model and with the help of the checkpoints, you'll be able to save the model at the end of each epoch.

```python
# Training our model and calculating its execution time

start = time.time()

history = model.fit_generator(train_generator, steps_per_epoch=steps_per_epoch, epochs=num_epochs, verbose=1, 
                    callbacks=callbacks_list, validation_data=val_generator, 
                    validation_steps=validation_steps, class_weight=None, workers=1, initial_epoch=0)

end = time.time()

print('Execution time for this model is',end-start,'seconds')
```

```python
# Performance Metrics

acc = history.history['categorical_accuracy']
val_acc = history.history['val_categorical_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(num_epochs)

font = {'size': 12}
plt.rc('font', **font)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Log Loss')

plt.subplots_adjust(wspace=0.5)

plt.show()
```
